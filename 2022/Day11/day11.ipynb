```python
from math import floor, lcm
NULL_STR = ""
```

```python
with open("input.txt") as file:
    lines = file.readlines()
```

```python
tokens = [line.strip().split(" ") for line in lines]
```

```python
class Monkey():

    def __init__(self, starting_items: list[int], operation: str, test: int, if_true: int, if_false: int):

        self.items = [int(item) for item in starting_items]
        self.operation = operation
        self.test = int(test)
        self.throw_to = [int(if_false), int(if_true)]
        self.times_inspected = 0
        self.lcm = None

    def inspect(self, monkeys: list, to_print=True):

        for item in self.items:
            if to_print:
                print(f"Monkey inspects an item with a worry level of {item}.")
            item = eval(self.operation)
            if to_print:
                print(f"Worry level is {self.operation} to {item}.")
            item = int(item % self.lcm)

            throw_ind = item % self.test == 0
            is_isnt = ["not ", NULL_STR]
            if to_print:
                print(
                    f"Current worry level is {is_isnt[throw_ind]}divisible by {self.test}.")
            target_monkey = self.throw_to[throw_ind]
            if to_print:
                print(
                    f"Item with worry level {item} is thrown to monkey {target_monkey}.")
            monkeys[target_monkey].items.append(item)
            self.times_inspected += 1

        self.items = []

    def __str__(self):
        string = "Monkey:\n"
        string += f"  items: {self.items}\n"
        string += f"  operation: {self.operation}\n"
        string += f"  test: {self.test}\n"
        string += f"  throw to: {self.throw_to}\n"
        string += f"  lcm: {self.lcm}\n"
        return string
```

```python
def parse_operation(operation_tokens):
    op = NULL_STR
    for token in operation_tokens:
        op += " "+token
    op = op.replace("old", "item")
    return op
```

```python
monkeys = []
for i in range(0, len(tokens), 7):
    monkey_tokens = tokens[i:i+6]
    starting_items = [int(x.replace(",", NULL_STR))
                      for x in monkey_tokens[1][2:]]
    operation = parse_operation(monkey_tokens[2][3:])
    test = monkey_tokens[3][3]
    if_true = monkey_tokens[4][5]
    if_false = monkey_tokens[5][5]
    monkeys.append(Monkey(starting_items, operation, test, if_true, if_false))

# Find the lowest common number of test numbers, to keep stress manageable
lowest_cm = 1
for monkey in monkeys:
    lowest_cm = lcm(lowest_cm, monkey.test)
for monkey in monkeys:
    monkey.lcm = lowest_cm
_ = [print(monkey) for monkey in monkeys]
```

```python
n_rounds = 10000
for round in range(n_rounds):
    # print(f"Round {round+1}:")
    for i, monkey in enumerate(monkeys):
        monkey.inspect(monkeys, False)
    if round % 1000 == 0:
        print(round)
        for i, monkey in enumerate(monkeys):
            print("Monkey " + str(i)+"\n")
            print(monkey)
```

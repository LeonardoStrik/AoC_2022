```python
# input: number of calories
# one item per line
# seperated by a blank line
# output: how many calories by the Elf carrying the most calories
```

```python
input = open("input", mode = 'r')
input = input.readlines()

for i in range(len(input)):
    input[i] = input[i][:-1]

sub = []
complete = []
input.append("")
for i in range(len(input)):
    if input[i] != '':
        sub.append(input[i])
    else:
        sub = [int(x) for x in sub]
        sum_sub = sum(sub)
        complete.append(sum_sub)
        sub = []

max(complete)
```

```python
# Find top three Elves
# Output: total calories of the top three
```

```python
complete.sort(reverse = True)
top_three = sum(complete[0:3])
top_three
```
